```python
# change to lower_case
def lower_case(text):
    
    return text.lower()

# change contraction words such sa I'm = I am, shouldn't = should not
def change_contractions(text):
    
    expanded_words = [contractions.fix(word) for word in text.split()]

    expanded_text = ' '.join(expanded_words)
    return expanded_text

# Remove Punctuations
def remove_punctuations(text):
  
  regex = re.compile('[^a-zA-Z0-9]')
  #first parameter is the replacement, second parameter is your input string
  new_text = regex.sub(' ', text);
  return new_text

# Remove extra whitespaces
def remove_extra_whitespace(text):
    return " ".join(text.split())
# Removing stopwords removes low-level information from our text in order to give more focus to the important information
def remove_stopwords(text):
    stopwords_list = stopwords.words('english')
    words = text.split() 
    clean_words = [word for word in words if (word not in stopwords_list) and len(word) > 1]
    return " ".join(clean_words)

def get_wordnet_pos(text):
    # Map POS tag to first character lemmatize() accepts
    tags = nltk.pos_tag(text)
    tag_dict = {"J": wordnet.ADJ,
                "N": wordnet.NOUN,
                "V": wordnet.VERB,
                "R": wordnet.ADV}
    tags = [tag_dict.get(tag[1][0],  wordnet.NOUN) for tag in tags]
    return tags

def lemmaSentence(text):
    lemmatizer = WordNetLemmatizer()
    lemma_text = ''
    tok_text = word_tokenize(text)
    tags = get_wordnet_pos(tok_text)
    for i in range(len(tok_text)):
        lemma_text = lemma_text + ' ' + lemmatizer.lemmatize(tok_text[i], tags[i])
    return lemma_text[1:] 

def remove_square_brackets(text):
    text = (text.strip("[]"))
    text = (text.strip("\'"))
    return text
    

# Create Word Cloud Visualization of each individual labels
def get_wordcloud(text, color='viridis'):
  # Create a WordCloud object
  wordcloud = WordCloud(background_color="white", max_words=5000, contour_width=3, contour_color='steelblue', colormap=color,collocations=False)
  # Generate a word cloud
  return wordcloud.generate(text)
```

```python
# Cleaning the provision column
def clean_text(data):
    clean_data = data.copy()
    clean_data['provision'] = clean_data['provision'].apply(lower_case)
    clean_data['provision'] = clean_data['provision'].apply(change_contractions)
    clean_data['provision'] = clean_data['provision'].apply(remove_punctuations)
    clean_data['provision'] = clean_data['provision'].apply(remove_stopwords)
    clean_data['provision'] = clean_data['provision'].apply(remove_extra_whitespace)
    clean_data['provision'] = clean_data['provision'].apply(lemmaSentence)
    
    clean_data['label'] = clean_data['label'].apply(remove_square_brackets)
    return clean_data
```

### Why Word2Vec

Answer: 


1. Word2Vec is able to capture the meaning of words in a continuous vector space, which makes it easy to perform mathematical operations on the embeddings, such as computing similarities and distances between words.

2. Word2Vec can handle out-of-vocabulary (OOV) words effectively by using sub-sampling and negative sampling techniques.

3. Word2Vec is computationally efficient and can be trained on large corpus of text.

4. Word2Vec has been extensively used and evaluated in a wide range of natural language processing tasks, and has been shown to be effective in many cases.

```python
W2V_SIZE = 300 # represents how big your dimensional size of your word2vec, 
W2V_WINDOW = 2 # Window_Size refers to the number of words on either side of a target word that are used to predict the target word
              # so if size = 5 , the window is (n-5),(n-4), (n-3)....n, (n+1)....(n+5)word
W2V_EPOCH = 32 # Number of epochs to train the word2vec
W2V_MIN_COUNT = 2 # This mean that, if the word that occurs less than 2 times, will be drop away from the dictionary

word2vec_model = Word2Vec(vector_size=W2V_SIZE, window=W2V_WINDOW, min_count=W2V_MIN_COUNT, workers=8)
```

```python
# Tokenize the provision columns
from nltk import word_tokenize, WordNetLemmatizer
tokens = [word_tokenize(text) for text in clean_dataset.provision]
tokens
```

```python
word2vec_model.build_vocab(tokens)
```

```python
word2vec_model.train(tokens, total_examples=len(tokens), epochs=32)
```

```python
# This shows the relationship of some words shared the same similaritiy as "provision", based on cosine similarity
word2vec_model.wv.most_similar("provision") 
```

```python
# See what is the longest string of text
text_length =  clean_dataset['provision'].str.len()
max_len = max(text_length)
max_len  
```

```python
from sklearn.preprocessing import LabelEncoder

# Prepare the data
le = LabelEncoder()
```

```python
# Now we want to split the dataset into train, validation and test in the proportion 80/10/10 respectively

# Split the data into train and validation sets
X_train, X_val, y_train, y_val = train_test_split(clean_dataset['provision'], clean_dataset['label'], test_size=0.2, random_state=42)

#Further split val data into val and test sets into half
X_val, X_test, y_val, y_test = train_test_split(X_val, y_val, test_size=0.5, random_state=42)
```

```python
test_csv = pd.DataFrame(X_test)
test_csv.to_csv("test_csv.csv")
```

```python
# label encoding for our categorical variable
y_train = le.fit_transform(y_train)
y_val = le.transform(y_val)
y_test = le.transform(y_test)
```

```python
# One-hot encode the target variable
y_train = to_categorical(y_train, num_classes)
y_val = to_categorical(y_val, num_classes)
y_test = to_categorical(y_test, num_classes)
```

```python
y_train
```

```python
tokenizer = Tokenizer()
# Tokenization helps to reduce the dimensionality of the data, making it easier to analyze and process.
tokenizer.fit_on_texts(clean_dataset.provision)

vocab_size = len(tokenizer.word_index) + 1
print("Total words", vocab_size) # Shows the vocacbulary size of the datasets
```

```python
# Pad the sequences to make all the same == maxlength
# By right when we doing padding our sequences, the maxlen argument must be maxlen of the longest in text
# But however, the longest in our dataset is 5152 and its too long....
# Therefore, to prevent the high dimensional Word2Vec from going into RAM memory overload, we only read first 1000 word

max_len = 1000
padded_X_train = pad_sequences(tokenizer.texts_to_sequences(X_train), maxlen=max_len)
padded_X_val = pad_sequences(tokenizer.texts_to_sequences(X_val), maxlen=max_len)
padded_X_test = pad_sequences(tokenizer.texts_to_sequences(X_test), maxlen=max_len)
```

```python
print(X_train)
```

```python
# Create embedding matrix
embedding_matrix = np.zeros((vocab_size, W2V_SIZE))
for word, i in tokenizer.word_index.items():
    if word in word2vec_model.wv:
        embedding_matrix[i] = word2vec_model.wv[word]
print(embedding_matrix.shape)
```

```python
# Creating the embedding layer based on the vocab size, Word2Vec Size, weights of the matrix as well lenth of the text
embedding_layer = Embedding(vocab_size, W2V_SIZE, weights=[embedding_matrix], input_length=max_len)
```

```python
# Creating a LSTM Model 
# Dropout is use to prevent the model from overfitting
# The activation function will be softmax since we are doing multi-class classification
# loss function we will be selecting categorical crossentropy

LSTM_model = Sequential()
LSTM_model.add(embedding_layer)
LSTM_model.add(Dropout(0.5))
LSTM_model.add(LSTM(128, dropout=0.2))
LSTM_model.add(Dense(num_classes, activation='softmax'))
LSTM_model.compile(optimizer='adam',loss='categorical_crossentropy', metrics=['accuracy'])
```

```python

LSTM_model.summary()
```

```python
# Callbacks to incurr earlystopping
from keras.callbacks import ReduceLROnPlateau, EarlyStopping

callbacks = [ ReduceLROnPlateau(monitor='val_loss', patience=3, cooldown=0),
              EarlyStopping(monitor='val_accuracy', min_delta=1e-4, patience=3)]
```

```python
history = {}
history["model"] = LSTM_model.fit(padded_X_train, y_train,
                batch_size=32,
                epochs=10,
                validation_data=(padded_X_val, y_val),
                verbose=1,
                callbacks=callbacks)
```

```python
# Saving the model
LSTM_model.save("LSTM_model")
```

```python
def plot_loss_accuracy_curve(history):
    fig, ax = plt.subplots(2,1, figsize=(15,15))
    fig.tight_layout(pad=5.0)
    ax[0].plot(history.history['accuracy'], label="Training")
    ax[0].plot(history.history['val_accuracy'], label="Validation")
    ax[0].set_title("Train Accuracy vs Validation Accuracy")
    ax[0].set_xlabel("Epoch")
    ax[0].set_ylabel("Accuracy")
    ax[0].legend()
    # Getting only the axes specified by ax
    extent = ax[0].get_window_extent().transformed(fig.dpi_scale_trans.inverted())
    # Saving it to a pdf file. 
    fig.savefig('figures/'+'accuracy.jpg', bbox_inches=extent.expanded(1.1, 1.2))
 
    
    ax[1].plot(history.history['loss'], label="Training")
    ax[1].plot(history.history['val_loss'], label="Validation")
    ax[1].set_title("Train Loss vs Validation Loss")
    ax[1].set_xlabel("Epoch")
    ax[1].set_ylabel("Loss")
    ax[1].legend()
    # Getting only the axes specified by ax
    extent = ax[1].get_window_extent().transformed(fig.dpi_scale_trans.inverted())
    # Saving it to a pdf file. 
    fig.savefig('figures/'+'loss.jpg', bbox_inches=extent.expanded(1.1, 1.2))
    plt.show()
```

```python
plot_loss_accuracy_curve(history["model"])
```

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score

# A helper function for creating Confusion Matrix and displaying the report
def Confusion_Matrix(y_true, y_pred, labels):
    
    # Create a confusion matrix
    conf_mat = confusion_matrix(y_true, y_pred, labels=labels)
                                                        
    plt.figure(figsize = (10,5))
    sb.set(font_scale=1.2)
    sb.heatmap(conf_mat, annot=True, cmap="YlGnBu", fmt='d', xticklabels=labels, yticklabels=labels)
    plt.xlabel("Predicted Label")
    plt.ylabel("True Label")
    plt.title("Confusion Matrix")
    plt.savefig("figures/ConfusionMatrix.jpg")
    plt.show()
    
    print(f"Classification Report:\n\n",classification_report(y_true, y_pred))
    print("-"*100)
```
